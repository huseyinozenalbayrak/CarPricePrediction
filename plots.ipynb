## Installing necessary packages

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("cleaned_data.csv")
```

## 1- Distribution of Car Prices

```python
# Set the aesthetic style of the plots
sns.set_style("whitegrid")

# Distribution of Car Prices
plt.figure(figsize=(10, 6))
sns.histplot(df['price'], kde=True, bins=30)
plt.title('Distribution of Car Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

## 2- Price vs. Kilometer (km) Scatter Plot

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='km', y='price', data=df)
plt.title('Price vs. Kilometer')
plt.xlabel('Kilometer')
plt.ylabel('Price')
plt.show()
```

## 3- Price vs. Year Scatter Plot

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='year', y='price', data=df)
plt.title('Price vs. Year')
plt.xlabel('Year')
plt.ylabel('Price')
plt.show()
```

## 4- Box Plot of Prices by Model

```python
plt.figure(figsize=(14, 8))
sns.boxplot(x='model', y='price', data=df)
plt.title('Box Plot of Prices by Model')
plt.xlabel('Model')
plt.ylabel('Price')
plt.xticks(rotation=45)
plt.show()
```

## 5- Box Plot of Prices by Fuel Type

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='fuel_type', y='price', data=df)
plt.title('Box Plot of Prices by Fuel Type')
plt.xlabel('Fuel Type')
plt.ylabel('Price')
plt.show()
```

## 6- Box Plot of Prices by City

```python
plt.figure(figsize=(14, 8))
sns.boxplot(x='city', y='price', data=df)
plt.title('Box Plot of Prices by City')
plt.xlabel('City')
plt.ylabel('Price')
plt.xticks(rotation=90)
plt.show()
```
